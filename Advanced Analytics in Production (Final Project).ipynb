# Step 2: Building the model in CPLEX

```python
from docplex.mp.model import Model
mdl=Model('PersonalScheduling')
```

### Initializing differents sets

```python
Numshift = [i for i in range(len(Shift['Shift']))] #Num of Workshift 
Numhour = [i for i in range(len(Shift.columns[1:]))] #Num of hour in the workshift
workShiftNode = [(i,j) for i in Numshift for j in Numhour] # defining the numer of shift and the hours

#Initialization of shift matrix
s=[]
for i in Numshift:
    for j in Numhour:
        s.append(Shift.iloc[i][j+1])
                 
workShift = {workShiftNode[k]: s[k] for k in range(len(s))}

#Initialization of cost matrix
costShift=[CostShift['Shift Cost'][i] for i in Numshift]

```

### decraring the decision variable

```python
X=mdl.integer_var_dict(Numshift,name = 'x')
```

```python
#Objective function

mdl.minimize(mdl.sum(X[i]*costShift[i] for i in Numshift))
```

```python
#demand must be satisfied
for j in Numhour:
    mdl.add_constraint(mdl.sum(X[k]*workShift[(k,j)] for k in Numshift)>=empleRequirement[0][j])

#full time workship must be assigne   
for i in range(7):
    mdl.add_constraint(X[i]>=1)
```

```python
print(mdl.export_to_string())
```

```python
solucion=mdl.solve(log_output=True)
```

```python
mdl.get_solve_status()
```

# Step 3: PL assign the work shift to employees

```python
from docplex.mp.model import Model
mdl2=Model('AssigPersonal')
```

```python
#definir numero de nodos
numEmployee = [i for i in range(len(Availability['Employee']))]
numEmployee
```

```python
Numshift = [i for i in range(len(Shift['Shift']))]
Numshift
```

```python
#Store the solution in a dictionary data structure.
sol= {i: int(X[i].solution_value) for i in Numshift if X[i].solution_value > 0.9}
print(sol)
```

```python
#create set of variables
availaSet = [(i,j) for i in numEmployee for j in sol]
print(availaSet)
```

```python
#create matrix for employee availability
#Here we read the worker disponibility
e=[]
for i in numEmployee:
    for j in Numshift:
        e.append(Availability.iloc[i][j+1])

availability = np.asarray(e,dtype= int) #Transform the list into array type  
availability = np.reshape(availability,(len(numEmployee),len(Numshift)))
print(availability)
```

```python
#create desicion variable
Y=mdl2.binary_var_dict(availaSet,name = 'Y')
```

```python
#Objective function
mdl2.maximize(mdl.sum(Y[i] for i in availaSet))
```

```python
#We must assined the number of employee determined in the solutions
for j in sol:
    mdl2.add_constraint(mdl2.sum(Y[(i,j)] for i in numEmployee)==sol[j])
```

```python
#every workshift must respect the employee disponibility
for i in numEmployee:
    mdl2.add_constraints(Y[(i,j)]<=100*availability[i][j] for j in sol)
```

```python
#We must assined the number of employee determined in the solutions
for i in numEmployee:
    mdl2.add_constraint(mdl2.sum(Y[(i,j)] for j in sol)<=1)
```

```python
print(mdl2.export_to_string())
```

```python
solucion2=mdl2.solve(log_output=True)
```
